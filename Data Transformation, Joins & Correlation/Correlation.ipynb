```
from pandas_datareader import wb
import pandas as pd
import warnings
warnings.simplefilter('ignore')
YEAR = 2013
GDP_INDICATOR = 'NY.GDP.MKTP.CD'
data = wb.download(indicator=GDP_INDICATOR, country='all', start=YEAR, end=YEAR)
data
```

```
gdp = data.reset_index()
gdp
```

```
LIFE_INDICATOR = 'SP.DYN.LE00.IN'
data = wb.download(indicator=LIFE_INDICATOR, country='all', start=YEAR, end=YEAR)
life = data.reset_index()
life.head(50)
```

```
gdp = gdp[47: ]
headings = ['country', GDP_INDICATOR]
gdp = gdp[headings]
gdp
```

```
gdp.dropna()
gdp.head(150)
```

```
life = life[47: ]
life = life.dropna()
headings = ['country', LIFE_INDICATOR]
life = life[headings]
life.head(50)
```

```
gdpVsLife = pd.merge(gdp, life, on='country', how='inner')
gdpVsLife.head()
```

```
def usdToGbp(usd):
  pound = usd / 1.564768
  return pound / 1000000
```

```
column = gdpVsLife[GDP_INDICATOR]
gdpVsLife['GDP mpounds'] = column.apply(usdToGbp)
gdpVsLife.head()
```

```
column = gdpVsLife[LIFE_INDICATOR]
life_expec = 'Life Expectancy (years)' 
gdpVsLife[life_expec] = column.apply(round)
gdpVsLife.head()
```

```
#Discarding the original columns and uisng the new columns for analysis.
headings = ['country', 'GDP mpounds', life_expec]
gdpVsLife = gdpVsLife[headings]
gdpVsLife.head(100)
```

```
gdpVsLife.dropna()
gdpVsLife.head(100)
```

```
#Spearman statistical coorelation. p-value =0.05.
from scipy.stats import spearmanr
gdpColumn = gdpVsLife['GDP mpounds']
lifeColumn = gdpVsLife[life_expec]
(correaltion, pvalue) = spearmanr(gdpColumn, lifeColumn)
print("Coorelation is", correaltion)
if pvalue < 0.05:
  print("Data is statistically significant")
else: 
  print("Data is not statistically significant")
```

```
#Here statistical computations don't provide accurate thus, for qualititive data
#we use plots, we are going to use a scatter plot.
gdpVsLife.plot.scatter(x="GDP mpounds", y=life_expec, grid=True, figsize=(15, 8), logx=True)
```
